# 7. Plot the avg tip by size  (lineplot)

```python
x=tips.groupby("size")["tip"].mean().index
x
```

```python
y=tips.groupby("size")["tip"].mean().values
```

```python
#functional
plt.figure(figsize=(8, 6))

plt.plot(x,y)
plt.xlabel("size");
```

```python
#oom
fig, ax = plt.subplots(figsize=(8,6))

ax.plot(x,y)
ax.set_xlabel("size");
```

# 8.Set the linestyle as "--"

```python
#oom
fig, ax = plt.subplots(figsize=(8,6))

ax.plot(x, y, linewidth=2.5, linestyle = "--" )
ax.set_xlabel("size");
```

# 10. Set an additional dimension using size column

```python
#with matplot
sns.set_style(style="darkgrid")
plt.scatter( x="tip",y='total_bill', data=tips , linewidth = 0.001, c = tips["size"], cmap="Spectral")
cbar=plt.colorbar()
plt.ylabel("total_bill")
cbar.set_label('size');
```

```python
#oom
ax = sns.scatterplot(y="total_bill", x="tip", hue="size", palette='viridis', data=tips)

norm = plt.Normalize(tips['size'].min(), tips['size'].max())
sm = plt.cm.ScalarMappable(cmap="viridis", norm=norm)
sm.set_array([])

# Remove the legend and add a colorbar
ax.get_legend().remove()
ax.figure.colorbar(sm, label="size")
ax.set(xlabel = "")
plt.show()
```

# 11. Make a box plot of total_bill column

# 12. Seperate the the boxplot above using size columns

```python
plt.figure(figsize=(8, 6))
sns.set_style(style="whitegrid")

tips.boxplot(column="total_bill", by="size");
```

# 13. Make the same plot using seaborn

```python
plt.figure(figsize=(8, 6))

sns.set_style(style="darkgrid")

sns.boxplot(data= tips,y="total_bill",x="size")
```

# 14. Make a violinplot instead of boxplot and discuss the difference between boxplot and violinplot

```python
plt.figure(figsize=(8, 6))

sns.set_style(style="darkgrid")

sns.violinplot(data= tips,y="total_bill",x="size")
```
